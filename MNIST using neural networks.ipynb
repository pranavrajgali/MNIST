Importing necessary libraries and tensorflow

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import seaborn as sn
```

1-> Loading the MNIST dataset

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[0]
```

2->Normalizing the data (*all the data is in the range 0-1 to make the learning process easier)*

```python
 X_train = X_train / 255
 X_test = X_test / 255
```

Visualisation of data (to check if the training data isnt corrupted or mislabelled)


```python
index = 90

plt.imshow(X_train[index], cmap=plt.cm.binary)
print(y_train[index])
```

```python
index = 15

plt.imshow(X_train[index], cmap=plt.cm.binary)
print(y_train[index])
```

3-> Flattening *(converting the images from 28x28(784 pixels) to a single row of 784 numbers)*

```python
X_train_flat = X_train.reshape(len(X_train), 28*28)
X_test_flat = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flat.shape
```

Building The neural network with 3 hidden layers (feedforward neural network)

*   Input 784 — Raw pixel data
*   Hidden 1 128 ReLU*(Learn basic patterns)*
*   Hidden 2 64  Sigmoid*(Learn complex patterns)*
*   Hidden 3 32  Sigmoid *(Refine patterns)*
*   Output 10    Softmax  *(Probabilities for digits 0-9)*


```python
model = keras.Sequential([
    keras.layers.Dense(128, input_shape=(784,), activation='relu'),
    keras.layers.Dense(64, activation='sigmoid'),
    keras.layers.Dense(32, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flat, y_train, epochs=10)
```

```python
model.evaluate(X_test_flat, y_test)
```

```python
y_pred = model.predict(X_test_flat)
y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
confusion_matrix = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_labels)
sn.heatmap(confusion_matrix, annot = True, fmt = 'd')
```

```python
import numpy as np
from sklearn.metrics import classification_report
# You'll use X_test_flat and y_test, which were loaded and processed earlier.

# 1. Generate the Raw Predictions and Convert to Digit Labels
# This step is essential to get the model's final guesses (0-9).
y_pred = model.predict(X_test_flat, verbose=0)
y_pred_labels = [np.argmax(i) for i in y_pred]

# 2. Get the Overall Success Rate (Accuracy)
# The model has this value saved from the training/evaluation phase.
# The result is [loss, accuracy]
loss, overall_accuracy = model.evaluate(X_test_flat, y_test, verbose=0)
print(f"Overall Success Rate (Accuracy): {overall_accuracy*100:.2f}%")

print("\n" + "="*50 + "\n")

# 3. Generate the Detailed Success Rates (Classification Report)
# This provides the breakdown of how well the model performed for each digit (0 through 9).
print("Detailed Success Rates (Classification Report):\n")
report = classification_report(y_test, y_pred_labels)
print(report)
```
